```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
## Display all the columns of the dataframe
pd.pandas.set_option('display.max_columns',None)
from pandas_profiling import ProfileReport
```

```python
df=pd.read_csv('movie.csv')
print(df.shape)
```

```python
df.head()
```

```python
df_cat=df.filter(['genres','original_language','production_companies','production_countries','actor','director'])
```

```python
df_cat.head()
```

```python
for i in df_cat.columns:
    print(len(df_cat[i].unique()))
```

### Make dummy variables in 'genres'

```python
df_cat['genres'].str.split(',', expand=True).stack().value_counts()
```

```python
dummies = df_cat['genres'].str.get_dummies(sep=',')
dummies=dummies.drop(dummies.columns[0], axis = 1) 
df_cat = pd.concat([df_cat, dummies], axis=1)
```

```python
df_cat.head()
```

### Make dummy variables with the most frequent variables in 'original_language'

```python
lang=df_cat.original_language.value_counts().sort_values(ascending=False)
lang.head(10)
```

```python
lang_7=list(lang.head(7).index) # we are taking most 7 unique languages
```

```python
lang_7
```

```python
for categories in lang_7:
    df_cat[categories]=np.where(df_cat['original_language']==categories,1,0)
```

```python
lang_7.append('original_language')
```

```python
df_cat[lang_7]
```

```python
df_cat['original_language'] = df_cat['original_language'].fillna(111)
df_cat.loc[(df_cat.original_language==111), ('en', 'fr', 'it', 'ja', 'de', 'es', 'ru')] = np.nan
```

### Make dummy variables with the most frequent variables in 'production_companies'

```python
df_cat['production_companies'].str.split(',', expand=True).stack().value_counts().head(15)
```

### we are taking most frequent 6 production_companies

```python
prod_com=list(['Warner Bros.','MGM','Paramount Pictures','Twentieth Century Fox Film Corporation','Universal Pictures','Columbia Pictures'])
```

```python
prod_com
```

```python
for categories in prod_com:
    df_cat[categories]=np.where(df_cat['production_companies'].str.contains(categories),1,0)
```

```python
prod_com.append('production_companies')
```

```python
df_cat[prod_com]
```

```python
df_cat['production_companies'] = df_cat['production_companies'].fillna(111)
df_cat.loc[(df_cat.production_companies==111), ('Warner Bros.',
 'MGM',
 'Paramount Pictures',
 'Twentieth Century Fox Film Corporation',
 'Universal Pictures',
 'Columbia Pictures')] = np.nan
```

### Make dummy variables with the most frequent variables in 'production_countries'

```python
prod_coun=df_cat['production_countries'].str.split(',', expand=True).stack().value_counts()
prod_coun.head(15)
```

```python
prod_coun_10=list(prod_coun.head(10).index) ## we are taking most frequent  10 production_countries
```

```python
prod_coun_10
```

```python
for categories in prod_coun_10:
    df_cat[categories]=np.where(df_cat['production_countries'].str.contains(categories),1,0)
```

```python
prod_coun_10.append('production_countries')
```

```python
df_cat[prod_coun_10]
```

```python
df_cat['production_countries'] = df_cat['production_countries'].fillna(111)
df_cat.loc[(df_cat.production_countries==111), ('United States of America','United Kingdom','France','Germany','Italy','Canada','Japan','Spain','Russia','India')] = np.nan
```

```python
df_cat.head()
```

```python
#dropping the columns those dummies have been created 
df_cat.drop(['genres','original_language','production_companies','production_countries'], axis=1, inplace=True) 
```

```python
df_cat['actor'].str.split(',', expand=True).stack().value_counts().head(20)
```

```python
df_cat['director'].str.split(',', expand=True).stack().value_counts().head(20)
```

```python
#We are dropping 'actor' and 'director', since they have many categories. It'll increases the features size a lot.
df_cat.drop(['actor','director'], axis=1, inplace=True) 
```
